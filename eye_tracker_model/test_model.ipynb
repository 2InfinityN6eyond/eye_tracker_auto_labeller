```python
import os
import glob
import time
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from datetime import datetime

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import random_split, DataLoader

from eye_tracker_dataset import EyeTrackerDataset
class EyeTrackerNet(nn.Module) :
    def __init__(self) :
        super(EyeTrackerNet, self).__init__()

        self.fc1 = nn.Linear(81 * 3, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32, 16)
        self.fc5 = nn.Linear(16, 8)
        self.fc6 = nn.Linear(8, 2)

    def forward(self, x) :
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        x = F.relu(x)
        x = self.fc4(x)
        x = F.relu(x)
        x = self.fc5(x)
        x = F.relu(x)
        x = self.fc6(x)
        x = F.relu(x)
        return x
```

```python
MODEL_PATH = "./checkpoints/model_20230428_105603_199"

model = EyeTrackerNet()
model.load_state_dict(torch.load(MODEL_PATH))
model.train(False)

SCREEN_SIZE = np.array([2055, 1328])

eye_tracker_dataset = EyeTrackerDataset()
```

```python

for idx in range(len(eye_tracker_dataset)) :
    input_data, label = eye_tracker_dataset[idx]

    label_np = label.detach().numpy()
    pred =  model(input_data)
    pred_np = pred.detach().numpy()

    pos_gt = (label_np * SCREEN_SIZE).astype(int)
    pos_pred = (pred_np * SCREEN_SIZE).astype(int)

    print(pos_gt, pos_pred)
```
